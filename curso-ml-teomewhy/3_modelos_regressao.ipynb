## Rela√ß√£o de Notas x Cervjas üç∫üìö

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("data/dados_cerveja_nota.xlsx")
df
```

```python
plt.style.use('dark_background')
plt.plot(df["cerveja"], df["nota"], 'o', color='#C74B50')
plt.grid(True, color='#282A3A')
plt.title("Rela√ß√£o Nota vs Cerveja")
plt.ylim(0, 11)
plt.xlim(0, 11)
plt.xlabel("Cerveja")
plt.ylabel("Nota")
ax = plt.gca()
ax.spines['top'].set_color('#282A3A')
ax.spines['bottom'].set_color('#282A3A')
ax.spines['left'].set_color('#282A3A')
ax.spines['right'].set_color('#282A3A')

plt.show()
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(df[["cerveja"]], df["nota"])
```

```python
a, b = reg.intercept_, reg.coef_[0]
print(f"a={a}; b={b}")
# a -> alocacao da reta
```

```python
X = df[["cerveja"]].drop_duplicates()
y_estimado = reg.predict(X)
y_estimado
```

```python
plt.plot(df["cerveja"], df["nota"], 'o', color='#C74B50')
plt.plot(X, y_estimado, '-', color='#B85C38')
plt.grid(True, color='#282A3A')
plt.title("Rela√ß√£o Nota vs Cerveja")
plt.ylim(0, 11)
plt.xlim(0, 11)
plt.xlabel("Cerveja")
plt.ylabel("Nota")
ax = plt.gca()
ax.spines['top'].set_color('#282A3A')
ax.spines['bottom'].set_color('#282A3A')
ax.spines['left'].set_color('#282A3A')
ax.spines['right'].set_color('#282A3A')
plt.show()
```

## Modelo de Decision Tree Regressor üå≥

```python
from sklearn import tree
arvore = tree.DecisionTreeRegressor(max_depth=2)
arvore.fit(df[["cerveja"]], df["nota"])
```

```python
y_estimado_arvore = arvore.predict(X)
y_estimado_arvore
```

```python
plt.plot(df["cerveja"], df["nota"], 'o', color='#C74B50')
plt.plot(X, y_estimado, '-', color='#B85C38')
plt.plot(X, y_estimado_arvore, '-', color='#FFD369')
plt.grid(True, color='#282A3A')
plt.title("Rela√ß√£o Nota vs Cerveja")
plt.ylim(0, 11)
plt.xlim(0, 11)
plt.xlabel("Cerveja")
plt.ylabel("Nota")
plt.legend(["Observa√ß√µes", "Regress√£o Linear", "√Årvore de Decis√£o"])
ax = plt.gca()
ax.spines['top'].set_color('#282A3A')
ax.spines['bottom'].set_color('#282A3A')
ax.spines['left'].set_color('#282A3A')
ax.spines['right'].set_color('#282A3A')
plt.show()
```
